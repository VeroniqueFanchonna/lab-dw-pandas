### Data Description

- Customer - Customer ID

- ST - State where customers live

- Gender - Gender of the customer

- Education - Background education of customers 

- Customer Lifetime Value - Customer lifetime value(CLV) is the total revenue the client will derive from their entire relationship with a customer. In other words, is the predicted or calculated value of a customer over their entire duration as a policyholder with the insurance company. It is an estimation of the net profit that the insurance company expects to generate from a customer throughout their relationship with the company. Customer Lifetime Value takes into account factors such as the duration of the customer's policy, premium payments, claim history, renewal likelihood, and potential additional services or products the customer may purchase. It helps insurers assess the long-term profitability and value associated with retaining a particular customer.

- Income - Customers income

- Monthly Premium Auto - Amount of money the customer pays on a monthly basis as a premium for their auto insurance coverage. It represents the recurring cost that the insured person must pay to maintain their insurance policy and receive coverage for potential damages, accidents, or other covered events related to their vehicle.

- Number of Open Complaints - Number of complaints the customer opened

- Policy Type - There are three type of policies in car insurance (Corporate Auto, Personal Auto, and Special Auto)

- Vehicle Class - Type of vehicle classes that customers have Two-Door Car, Four-Door Car SUV, Luxury SUV, Sports Car, and Luxury Car

- Total Claim Amount - the sum of all claims made by the customer. It represents the total monetary value of all approved claims for incidents such as accidents, theft, vandalism, or other covered events.

```python
# Identify numerical and categorical columns
# Professional tip: Always check data types to avoid calculation errors later
print(df.info())

# Check for null values
null_counts = df.isnull().sum()
print(f"Missing values per column:\n{null_counts}")
print(f"Audit : Nombre de lignes après nettoyage : {len(df)}") # Doit afficher ~1071
```

```python
df.rename(columns={'GENDER': 'Gender'}, inplace=True)
df.rename(columns={'ST': 'State'}, inplace=True)
#*In education, "Bachelors" could be replaced by "Bachelor"*
df.loc[df["Education"] == "Bachelors", "Education"] = "Bachelor"
#Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
df.loc[df["Gender"] == "F", "Gender"] = "Female"
df.loc[df["Gender"] == "Femal", "Gender"] = "Female"
df.loc[df["Gender"] == "female", "Gender"] = "Female"
df.loc[df["Gender"] == "M", "Gender"] = "Male"
#State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
df.loc[df["State"] == "AZ", "State"] = "Arizona"
df.loc[df["State"] == "Cali", "State"] = "California"
df.loc[df["State"] == "WA", "State"] = "Washington"
df['State'] = df['State'].fillna("Unknown")
#In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*
df.loc[df["Vehicle Class"] == "Sports Car", "Vehicle Class"] = "Luxury"
df.loc[df["Vehicle Class"] == "Luxury SUV", "Vehicle Class"] = "Luxury"
df.loc[df["Vehicle Class"] == "Luxury Car", "Vehicle Class"] = "Luxury"
df.loc[df["Vehicle Class"] == "SUV", "Vehicle Class"] = "Luxury"

df.columns = [col.lower().replace(" ", "_") for col in df.columns]

list(df.columns)
```

```python
df.head()
```

```python
df.loc[:, 'customer_lifetime_value'] = (
    df['customer_lifetime_value']
    .astype(str)                        # force the texte
    .str.replace('%', '', regex=False)  # supprenss %
    .astype(float)        # back to float
)

df.loc[:, 'total_claim_amount'] = (
    df['total_claim_amount']
    .astype(str)                        # force the texte
    .astype(float)                     # back to float
)
df.loc[:, 'monthly_premium_auto'] = (
    df['monthly_premium_auto']
    .astype(str)                        # force the texte
    .str.replace(',', '.').str.strip()
)
df['monthly_premium_auto'] = pd.to_numeric(df['monthly_premium_auto'], errors='coerce')  # back to float

print(df.info())
```

```python
df['state'] = df['state'].fillna("Unknown")
df['state'] = df['state'].fillna("Undefined")
df.head()
```

```python
# 1. On retire les espaces et on remplace la virgule par un point
df['monthly_premium_auto'] = df['monthly_premium_auto'].astype(str).str.replace(',', '.').str.strip()

# 2. On convertit en numérique (les erreurs deviennent des NaN)
df['monthly_premium_auto'] = pd.to_numeric(df['monthly_premium_auto'], errors='coerce')

# 3. On vérifie le type
print(f"Type de monthly_premium_auto: {df['monthly_premium_auto'].dtype}")
```
